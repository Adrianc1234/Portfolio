In this problem your task is to open and explore the NOAA weather data using Pandas. The data file name is [data/6153237444115dat.csv](data/6153237444115dat.csv). Please answer the given tasks write your code into the codeblocks below.

- **import Pandas module** (1 points)
- **read the data** using pandas into a variable called **`data`** (2 points)

**Important**: When reading the data, you need to tell Pandas that no-data values are specified with varying number of \* characters. You can do this by specifying a following parameter in the `read_csv()`-function: 
  
  `na_values=['*', '**', '***', '****', '*****', '******']`
     

```python
"""
we read the file.csv without some symbols.
"""
import pandas as pd
data= pd.read_csv('6153237444115dat.csv', sep=',',na_values=['*', '**', '***', '****', '*****', '******'] )
print(data)
```

In the codeblock below, print answers to these questions:
- How many rows is there in the data?
- What are the column names?
- What are the datatypes of the columns?

```python
"""
we use len to check the large of the data base.
we use .columns function to check the names of the rows
we use type function to see what type of data is there are in the rows.
"""
# Print number of rows in the dataframe:
print("rows in data:\n",len(data))

# Print column names:
print("names of the columns:\n ", data.columns)

# Print column datatypes:
print("data types:\n", type(data))


```

**Explore basic statistics in the data** (2 points)
- What is the mean Fahrenheit temperature in the data? (`TEMP` column)
- What is the standard deviation of the Maximum temperature? (`MAX` column)
- How many unique stations exists in the data? (`USAF` column)

Please use the suggested variable names to store the results

```python
# YOUR CODE HERE
"""
we use the .mean function to calculate the mean of the values in a specific row.
we make the same but using the .std to calculate the standar deviation.
we use unique to check the values that are uniques on the row
"""
# Calculate mean temperature
temp_mean=data['TEMP'].mean()
# Calculate standard deviaton of the maximum temperature:
temp_max_std =data['MAX'].std()
#Calculate number of unique stations:
station_count =len(data['USAF'].unique())
```

Once you have defined the variables above, you can print the answers using these print statements (you don't need to change anything in the code block below!)

```python
"""
we print the answers.
"""
print("Mean temperature in Fahrenheit is ", round(temp_mean,1))
print("Sandard deviation of maximum temperature is ", round(temp_max_std, 1))
print("The number of unique stations is ", station_count)
```
