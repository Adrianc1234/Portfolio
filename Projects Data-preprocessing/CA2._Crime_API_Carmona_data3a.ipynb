# STEP 1 LIST OF THE CRIME

```python
#we make the function to get the data
import requests
import json
import pandas as pd
#obtain a list
def crime():
    url = f'https://hoyodecrimen.com/api/v1/crimes'
    res = requests.get(url)
    results = json.loads(res.content)
    return results
```

```python
#we use the function
crime=crime()
```

```python
#we read and make the table
crime = pd.read_json(json.dumps(crime['rows']), orient = 'records')
crime
```

# STEP 2- ALL QUADRANTS

```python
#we make the function to get these data
import requests
import json
import pandas as pd
#obtain a list
def qua():
    url = 'https://hoyodecrimen.com/api/v1/cuadrantes'
    res = requests.get(url)
    results = json.loads(res.content)
    return results
#we call the function
qua=qua()
```

# STEP 3- SEARCHING CRIME WITH A COORDENATE

```python
# we make a function to get the data
def coor(latitude, longitude):
    url = f'https://hoyodecrimen.com/api/v1/latlong/crimes/all/coords/{longitude}/{latitude}/distance/500'
    res = requests.get(url)
    results = json.loads(res.content)
    return results

longitude='-99.133334'
latitude='19.433735'
#we call the function and the variables.
coor=coor(latitude,longitude)
```

```python
#we read in table
coor = pd.read_json(json.dumps(coor['rows']), orient = 'records')
coor.head()
```

# STEP 4 - TOP OF CRIME OF 2016, 2017, 2018.

```python
#we get the data with the function TOP.
def TOP(start_date,end_date):
    url = f'https://hoyodecrimen.com/api/v1/cuadrantes/crimes/all/top/counts?start_date={start_date}&end_date={end_date}'
    res = requests.get(url)
    results = json.loads(res.content)
    return results

#we read the first year
start_date='2016-01'
end_date='2017-12'
crimes_2016=TOP(start_date,end_date)

#we read the second year
start_date='2017-01'
end_date='2018-12'
crimes_2017=TOP(start_date,end_date)

#We read the third year
start_date='2018-01'
end_date='2018-12'
crimes_2018=TOP(start_date,end_date)
#we read and make a data frame
crimes_2016= pd.read_json(json.dumps(crimes_2016['rows']), orient = 'records')
crimes_2017= pd.read_json(json.dumps(crimes_2017['rows']), orient = 'records')
crimes_2018= pd.read_json(json.dumps(crimes_2018['rows']), orient = 'records')

#we use the function .concat to concatenate the dataframes
data=pd.concat([crimes_2016,crimes_2017,crimes_2018])
```

# STEP 5- TABLE IN ALL SECTORS

```python
#we make a function to get the data
def sectors(start_date,end_date):
    url = f'https://hoyodecrimen.com//api/v1/sectores/all/crimes/all/series?start_date={start_date}-01&end_date={end_date}-12'
    res = requests.get(url)
    results = json.loads(res.content)
    return results

        
```

```python
#we run the function for 2016
start_date='2017'
end_date='2017'                                                                                                                   
data_2016=sectors(start_date, end_date)
data_2016= pd.read_json(json.dumps(data_2016['rows']), orient = 'records')
#for 2017
start_date='2017'
end_date='2017'                                                                                                                   
data_2017=sectors(start_date, end_date)
data_2017= pd.read_json(json.dumps(data_2017['rows']), orient = 'records')

#for 2018
start_date='2018'
end_date='2018'                                                                                                                   
data_2018=sectors(start_date, end_date)
data_2018= pd.read_json(json.dumps(data_2018['rows']), orient = 'records')
#we concatenate the three dataframes in one
data_table=pd.concat([data_2016,data_2017,data_2018])
#we print the new table.
data_table
```
