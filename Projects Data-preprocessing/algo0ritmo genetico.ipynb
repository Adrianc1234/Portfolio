```python
import numpy as np
import random
```

```python
#genetic alforitm variables
target = "universidad politecnica de yucatan"
population = 10000
treshold_mutation = 0.10
tershold_worst = 0.80
tershold_parent = 0.10
```

```python
#variable initialitation
letters=" qwertyuiopasdfghjklñzxcvbnm"
```

```python
#inialize population
words = []
for j in range(population):
    word = ""
    for i in range(len(target)):
        word += random.choice(letters)
    words.append([0,word])
```

```python
words
```

```python
def fitness(words,target):
    for entry in words:
        word = entry[1]
        score = 0
        for i in range(len(word)):
            if word[i]==target[i]:
                score+=1
        entry[0]=score
```

```python
fitness(words,target)
```

```python
def crossover(parent1,parent2):
    child=""
    for i in range(len(parent1)):
        choice=random.random()
        if choice > 0.5:
            child += parent1[i]
        else:
            child+= parent2[i]
    return child
```

```python
index = int(population*tershold_parent)
parents = words[:index]
```

```python
index = int(population*tershold_worst)
for i in range(index,population):
    parent_1 = random.choice(parents)
    parent_2 = random.choice(parents)
    child = crossover(parent_1[1],parent_2[1])
    words[i][1]=child
```

```python
def mutation(word):
    choice = random.random()
    if choice < treshold_mutation:
        index= random.randint(0,len(word)-1)
        word = word[:index]+ random.choice(letters)+word[index+1:]
    return word
```
