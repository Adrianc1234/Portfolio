### BILLBOARD

```python
# we read the data frame and  we put it in the data_3 variable.
data_3=data = pd.read_csv('billboard.csv', encoding= 'unicode-escape')
data_3=pd.DataFrame(data_3)
data_3
```

For the this part of the code, we declared a variable with all titles that i wanted to put for the new data frame. After that we use the titles for the melting process and in this step we used the function melt to change the positions of the variables, but although that we made that we need to change some things, because in the case of week, we need to change those values by int but we had a serie, then we needed to use the .extract function to get all values of the serie and we needed to make the convertition of "rank" column because we had to make a operation to get the new values of date column, then once that we already made that we can give the new titles for the new data frames and order it.

```python
# we wrote the titles for the new data frames.
titles = ["year",
           "artist.inverted",
           "track",
           "time",
           "genre",
           "date.entered",
           "date.peaked"]
#changing the position using the new titles.
data_3 = pd.melt(data_3, id_vars=titles, var_name="week", value_name="rank")
#we make a new column called "week" for the dataframe
data_3["week"] = data_3['week'].str.extract('(\d+)', expand=False).astype(int)# we get the values in a series with extract function and convert these in int
data_3["rank"] = data_3["rank"].astype(float)# we conver the values of "rank" to float
data_3 = data_3.dropna()#we eliminate NAN
data_3['date'] =  pd.to_datetime(data_3['date.entered']) + pd.to_timedelta((data_3['week'] - 1) * 7, 'd')#we make a formula to get the new date 
#with help of the others columns

# we hold the data of this column and we add the new title that is "date" for the dataframe
data_3 = data_3[["year", 
         "artist.inverted",
         "track",
         "time",
         "genre",
         "week",
         "rank",
         "date"]]
# we use sort function to order the data in an ascending way 
data_3= data_3.sort_values(ascending=True, by=["year","artist.inverted","track","week","rank"])
data_3.head()
```

### TB DATA

```python
# we only read againg the file and make a dataframe with it.
data_5= pd.read_csv('tb.csv')
data_5= pd.DataFrame(data_5)
data_5.head()
```

```python
#we make the melting process holding iso2 and year columns and using the data of cases and sex age.
datacd = pd.melt(data_5, id_vars = ['iso2', 'year'], value_name="cases", var_name="sex_age")
# we get the age limit upper and age limit lower.
age_data = datacd["sex_age"].str.extract("(\D)(\d+)(\d{2})")    
# we give the names of the columns
age_data.columns = ["sex", "age_low", "age_up"]
# we make a range of ages based in the lower and upper ages.
age_data["age"] = age_data["age_low"] + "-" + age_data["age_up"]
#we only use the function concat to combinate the data frames.
datacd = pd.concat([datacd, age_data], axis=1)
# we use drop to delete the unnecessary columns.
datacd = datacd.drop(['sex_age',"age_low","age_up"], axis=1)
datacd = datacd.dropna()#we use this fucntion to delete the data NAN
datacd = datacd.rename(columns={'iso2':'country'}) #we only change the name of the clumns.
datacd = datacd.sort_values(by =["country", "year", "sex", "age"], ascending = True) #like in the above code, 
#we order the values and use .reset_index to delete the index of the beginning and change by the other.
datacd.reset_index(drop = True, inplace = True)
datacd.head()# print.
```

### WEATHER DATA

```python
data_6= pd.read_csv('weather.txt', sep="\t")# we read the data frame with a specific separator.
data_6.head()
```

```python
#melting
data_6 = pd.melt(data_6, id_vars=['id', 'year', 'month', 'element'],
             value_vars=list(data_6.columns)[4:],
             var_name='date', value_name='value')
#we delete the NAN variables
data_6= data_6.dropna()
```

```python
#we change the string of date column by int
data_6['date'] = data_6['date'].str[1:].astype(int)

#we apply the format for the date with lambda function
data_6['date'] = data_6[['year', 'month', 'date']].apply(
    #02d formats an integer (d) to a field of minimum width 2 (2), with zero-padding on the left (leading 0):
    # we make the lambda with 4d to the year 02 for the month and 02 for the day with the string format
    lambda row: '{:4d}-{:02d}-{:02d}'.format(*row),
    axis=1)
```

```python
# we loc all the data with .loc function indicating the coordinates
data_6 = data_6.loc[data_6['value'] != '---', ['id', 'date', 'element', 'value']]
#Set the index to become the ‘id’, 'date','element' columns
data_6 = data_6.set_index(['id', 'date', 'element'])
#we order it as a stack
data_6 = data_6.unstack()
#we make a list to read in a rigth way the dataframe and we help using .get function. 
data_6.columns = list(data_6.columns.get_level_values('element'))
data_6 = data_6.reset_index()
data_6.head()
```
