```python
# Library import
import pandas as pd
import numpy as np
```

I use for this part `pandas` to read each data set but I wrote the separator like: 

`sep = '\s+'` saying that the separation is a white space.

```python
data_1= pd.read_csv('weather01.txt', sep = '\s+')
data_2= pd.read_csv('weather02.txt', sep = '\s+')
data_3= pd.read_csv('weather03.txt', sep = '\s+')
data_4= pd.read_csv('weather04.txt', sep = '\s+')
data_5= pd.read_csv('weather05.txt', sep = '\s+')
data_6= pd.read_csv('weather06.txt', sep = '\s+')
```

In this part I make a group of data sets that I want to join in a new variable called `data_finalsets` where I select all the data, after that I used `.concat()` to make a concatenation of the six dataframes, without sort it.

The second step was delete the unnecesary column called `X` and I used `.drop()` to do this.

After that I applied a `.melt()` to make other structure for my data frame, in this case i used the columns 'year', 'month' and 'measure' like id or statics column, and I said that my other columns will be the values for the new column called `day` and obviously for each value that each column had will be in other column called `value`.


```python
data_finalsets = [data_1, data_2, data_3, data_4, data_5, data_6]# we make a set of all data frame.

# concatenation
data_final = pd.concat(data_finalsets, axis=1, sort=False)# we joing all data frames.
data_final.head(25)

# delete the X column that is unnecesary.
data_final = data_final.drop('X', axis=1)
# using Melt function
data_final = pd.melt(data_final, id_vars=['year', 'month', 'measure'], 
             var_name='day', value_name='value') # we make a melt to reorder the data frame.

data_final.head()# I print the data frame
```

In this part I Used `.string[1:].astype('int')` to delete all the string with the leter `X` on the column and only keep the integer value.

```python
# I deleted the string X for each value in day and only keep the Int value
data_final['day'] = data_final['day'].str[1:].astype('int')
data_final.head() # we print the data frame
```

In this step I made a tranformation using the time format using the three columns `year`,`month` and `year` to make a new column with the date but in a better format.

```python
# create the variable date
# use a lambda function to obtain the date with format yyyy-mm-dd
data_final['date'] = data_final[['year', 'month', 'day']].apply(
    lambda row: '{:4d}-{:02d}-{:02d}'.format(*row),    # we use lambda fucntion to make a format to these date.
    axis=1)
data_final.head()
```

In this step I used `.drop` and `.dropna` function to delete all the colums that we joined in the new date column, in this case they are `year`, `month` and `day`. By the other hand I make a condition to save only the columns that do no thave NA values inside of them, but I saw that actually the values were not Na values, they are strings with this format `data_final[data_final['value']!='<NA>']` to get only data that are full.

```python
# drop unnecesary columns
data_final = data_final.drop(['year',"month","day"], axis=1)
data_final.value.dropna()
data_final = data_final[data_final['value']!='<NA>']
data_final.head()
```

In this part I made a pivot using like index the date column and my new columns were the values inside of mesure column. I used the function `.pivot_table()`

```python
# Unmelting column "measure" with pivot
data_final = data_final.pivot_table(index=['date'], columns=["measure"], values=["value"], margins = False, aggfunc='first')
data_final.reset_index(drop=False, inplace=True)
```

Finally in this part I applied a `.dropna()` to delete the rest Na values and only get full columns.

```python
data_final = data_final.dropna()
data_final.head()
```
